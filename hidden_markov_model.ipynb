```python
%tensorflow_version 2.x
import tensorflow_probability as tfp
import tensorflow as tf
```

```python
tfd = tfp.distributions
initial_distribution = tfd.Categorical(probs=[0.8, 0.2]) # refers to the first day being 80% cold
transition_distribution = tfd.Categorical(probs=[[0.7, 0.3], [0.2, 0.8]])
observation_distribution = tfd.Normal(loc=[0., 15.], scale=[5., 10.])
# location defines the mean and scale defines the standard deviation
```

```python
model = tfd.HiddenMarkovModel(
    initial_distribution=initial_distribution,
    transition_distribution=transition_distribution,
    observation_distribution=observation_distribution,
    num_steps = 7
)
```

```python
mean = model.mean()
# calculates the probabilities, partially calculated tensors

with tf.compat.v1.Session() as sess:
  print(mean.numpy())
```
